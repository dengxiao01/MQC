```python
import pandas as pd
import os
import subprocess
```

```python
slurm_head = """#!/bin/bash
#SBATCH --job-name=hostname
#SBATCH --partition=qcpu_23i   
#SBATCH -N 1
#SBATCH -n 1
#SBATCH --mem=4G  #
#SBATCH --mail-type=end
#SBATCH --mail-user=dengxg@tib.cas.cn
#SBATCH --output=/hpcfs/fhome/dengxg/xml0321/%j.out
#SBATCH --error=/hpcfs/fhome/dengxg/xml0321/%j.err

conda activate dxbase

cd /hpcfs/fproject/dengxg/mqc
"""
```

```python
#  构建模型输入输出

model_input_dir = '/hpcfs/fproject/dengxg/mqc/mqc/local_test_data/embl_gems/'
model_output_dir = '/hpcfs/fproject/dengxg/mqc/tmp/web_CARVEME_COMMEN/'
script_dir = '/home/dengxiao/mqc/mqc/slurm_script/'
model_file = '/home/dengxiao/mqc/mqc/local_test_data/embl_gems/model_list.tsv'
models = pd.read_csv(model_file,sep='\t').file_path.to_list()
models_input = [model_input_dir+item for item in models]
models_output = [model_output_dir+ item.split('/')[-1].split('.')[0] for item in models_input]

```

```python
# 生成脚本文件
for i in range(len(models_input)):
    run_script = f"python start3.py -m '{models_input[i]}' -o '{models_output[i]}'"
    run_script = slurm_head + run_script
    with open(script_dir+models_input[i].split('/')[-1].split('.')[0]+'.slurm', 'w') as file:
        file.write(run_script)

    # print(run_script)
```

```python
# 提交脚本
all_files = os.listdir(script_dir)
slurm_files = [script_dir+file for file in all_files if file.endswith(".slurm")]
```

```python
slurm_files
```

```python
slurm_filescommand = "ls"
for item in slurm_files:
    result = subprocess.run(f'sbatch {item}', shell=True, stdout=subprocess.PIPE, text=True)
```

```python
import concurrent.futures
import threading
from functools import partial

import sys
sys.path.append('/home/dengxiao/mqc/')
from start3 import main


program_arguments = [(k,v) for k in models_input for v in models_output]
program_arguments
with concurrent.futures.ThreadPoolExecutor() as executor:
    # 使用 functools.partial 来传递参数
    futures = [executor.submit(partial(main, *args)) for args in program_arguments]

    # 获取结果
    results = [future.result() for future in concurrent.futures.as_completed(futures)]

# 处理结果
for args, result in zip(program_arguments, results):
    print(f"Program with args {args} result: {result}")
```
