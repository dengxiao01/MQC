```python
import os
import pandas as pd 
import time
from openpyxl import Workbook, load_workbook
import fcntl

def append_to_excel_with_retry(excel_filename, data, max_retries=10, retry_interval=1):
    retries = 0
    headers=['model','NADH','NADPH','FADH2','FMNH2','Q8H2','MQL8','DMMQL8','ATP','CTP','GTP','UTP','ITP','metabolite','yield','biomass','carbon source supply','restricted metabolites']
    while retries < max_retries:
        try:
            # 尝试获取文件锁
            with open(excel_filename, 'a') as file:
                fcntl.flock(file, fcntl.LOCK_EX | fcntl.LOCK_NB)        
                # 读取现有的 Excel 文件
                try:
                    df = pd.DataFrame([data],columns=headers)
                    # 尝试读取现有文件
                    try:
                        existing_df = pd.read_excel(excel_filename, sheet_name='Sheet1')
                        # 将新的数据追加到现有数据框
                        existing_df = existing_df.append(df, ignore_index=True)
                    except FileNotFoundError:
                        # 如果文件不存在，直接使用新数据框
                        existing_df = df
                    # 将数据框写入Excel文件
                    existing_df.to_excel(excel_filename, index=False, header=True, sheet_name='Sheet1')

                    return  # 写入成功，退出循环
                except Exception as e:
                    print(f"Error while processing Excel file: {e}")
                finally:
                    # 释放文件锁
                    fcntl.flock(file, fcntl.LOCK_UN)
        except BlockingIOError:
            print(f"Could not get lock for {excel_filename}. Retrying...")
            retries += 1
            time.sleep(retry_interval)
    print(f"{data} Exceeded maximum retries. Could not write to {excel_filename}.")

# 测试
excel_filename = '/home/dengxiao/mqc/tmp/control定量分析.xlsx'

# 模拟多个进程追加到 Excel 文件
for i in range(5):
    data = [f'{i}', '', '', '', '', '', '', '', '', '', '', '', '', '', '', '', ['Ex_glc_D : (-10, 1000)'], {}]
    append_to_excel_with_retry(excel_filename, data)
    time.sleep(1)  # 模拟一些处理时间
```
